```python
from IPython.display import clear_output
import time

def display_board(board):
    clear_output()
    print("\t|\t|\t\n    {}   |   {}   |    {}    \n\t|\t|\t\n--------+-------+--------".format(board[1],board[2],board[3]))
    print("\t|\t|\t\n    {}   |   {}   |    {}    \n\t|\t|\t\n--------+-------+--------".format(board[4],board[5],board[6]))
    print("\t|\t|\t\n    {}   |   {}   |    {}    \n\t|\t|\t".format(board[7],board[8],board[9]))
    
def board_def():
    
    for i in range(1,10):
        board[i] = ' '
    display_board(board)
    
def player_marker():
    while(True):
        player1 = input("select you marker- either 'X' or 'O' \n").upper()
        if (player1 == 'X'):
            player2 = 'O'
            break
        elif (player1 == 'O'):
            player2 = 'X'
            break
        
    return (player1,player2)

def enter_position(player):      
    position = int(input(' Player {} enter available position from 1-9\n'.format(player)))
    while(board[position] == 'X' or board[position] == 'O'):
        position = int(input(' Player {}, {} position already occupied please enter another\n'.format(player,position)))
    board[position] = player
    display_board(board) 
        
    
def game_win(board):
    if( board[1] == board[2] == board[3] != ' '):
        return board[1]
    elif(board[4] == board[5] == board[6] != ' '):
        return board[4]
    elif(board[7] == board[8] == board[9] != ' '):
        return board[7]
    elif(board[1] == board[4] == board[7] != ' '):
        return board[1]
    elif(board[2] == board[5] == board[8] != ' '):
        return board[2]
    elif(board[3] == board[6] == board[9] != ' '):
        return board[3]
    elif(board[1] == board[5] == board[9] != ' '):
        return board[1]
    elif(board[3] == board[5] == board[7] != ' '):
        return board[3]
    else:
        return None
        
def check_board(board):
    empty = ' ' in board.values()
    return empty
            
player1,player2 = player_marker() 
print("{} will go first".format(player1))
board = dict()
board_def()
turn = player1
while(True):
    if turn == player1:
        enter_position(player1)
        turn = player2
    else:
        enter_position(player2)
        turn = player1
    winner = game_win(board)
    if(winner!= None):
        print('{} Wins!!!!!'.format(winner))
        break
    else:
        empty = check_board(board)
        if empty == True:
            continue
        else:
            win = game_win(board)
            if win == None:
                print('Match draws - No one Wins')
                break
            
    
        
time.sleep(5)
clear_output()
```
